```python
from si.data.dataset import Dataset
import numpy as np

# create a dataset
X = np.random.rand(150, 30)
y = np.random.randint(0, 2, 150)
dataset = Dataset(X, y)
dataset.shape()
```

```python
# get the X
dataset.X
```

```python
#get the features names
dataset.features
```

```python
# has label?
dataset.has_label()
```

```python
# print the label
dataset.y
```

```python
# get classes
dataset.get_classes()
```

```python
# statistics summary
dataset.summary()
```
